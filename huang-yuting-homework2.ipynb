# 1. pig 

```python
pigdata.shape # three channels(R=red, G=green, B=blue) and no opacity
# 300 *144 = 43200 (all the pixels)
```

```python
plt.imshow(pigdata)
```

```python
np.unique(pigdata) # unique numbers in pigdata
```

```python
pigdata.reshape(-1, pigdata.shape[2])
```

```python
pigdata.reshape(-1,pigdata.shape[2]).shape # all the pixels
```

```python
np.unique(pigdata.reshape(-1, pigdata.shape[2]), axis = 0) # unique values for all the pixels 
```

```python
fix, ax = plt.subplots(2, 2, figsize = (15, 15))# we have 2 rows and 2 columns

ax[0, 0].hist(pigdata[:, :, 0].flatten(), bins = 256, color = 'red')# [0,0]upper and left, 0 = Red channel
ax[0, 1].hist(pigdata[:, :, 1].flatten(), bins = 256, color = 'green')# [0,1]upper and right, 1 = Green channel
ax[1, 0].hist(pigdata[:, :, 2].flatten(), bins = 256, color = 'blue')# [1,0]lower and left, 2 = Blue channel

# label x-axis
ax[0, 0].set_xlabel('Red channel bins')
ax[0, 1].set_xlabel('Green channel bins')
ax[1, 0].set_xlabel('Blue channel bins')

for i in range(2):
    for i in range(2):
        ax[i, i].set_ylabel('# of pixels')
    
ax[1, 1].imshow(pigdata)# [1,1]lower and right
plt.show()
```

I make 2 rows and 2 columns histograms because pixels are distributed in a wide range. Thus, it helps to observe the RGB color space.
Three channels have many pixels are 255 because the white color occupies a big portion of the image.

```python
fig = plt.figure(figsize=(15,10))
gs = fig.add_gridspec(3,2) # three rows (3 color histograms), 2 columns (histograms on the left, fig on right)

ax1 = fig.add_subplot(gs[0,0]) # upper left, plot red
ax1.hist(pigdata[:,:,0].flatten(),bins=256, color='red')
ax1.set_xlabel('R channel')

ax2 = fig.add_subplot(gs[1,0]) # middle row (1), first column (0) -- green
ax2.hist(pigdata[:,:,1].flatten(),bins=256,color="green")
ax2.set_xlabel('G channel')

ax3 = fig.add_subplot(gs[2,0]) # last row (2), first column (0) -- blue
ax3.hist(pigdata[:,:,2].flatten(),bins=256,color="blue")
ax3.set_xlabel('B channel')

ax4 = fig.add_subplot(gs[:,1]) # all rows (:), 2nd column (1) for the image
ax4.imshow(pigdata)

plt.show()
```
