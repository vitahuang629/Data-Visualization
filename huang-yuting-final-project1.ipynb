# 2. Summerize the characteristics 

### What does it represent?

The dataset represents the data on adult's diet, physical activity and weight status from Behavioral Risk Factor Surveillance System. This data is composed by national and state data on Fruits and Vegetables, Obesity / Weight Status, and Physical Activity.

### What are the fields / columns / rows?

1. YearStart > Year start
2. YearEnd > Year End
3. LocationAbbr > Location abbreviation
4. LocationDesc > Location description
5. Datasource > Name of Data Source
6. Class > Class Description
7. Topic > Topic Description
8. Question > Question Description
9. Data_Value_Unit > Description of unit, ex:%(percent)
10. Data_Value_Type > Description of type of data, ex: Value Percentage, Number
11. Data Value > Data value, ex: percentage and text
12. Data_Value_Alt > Numeric representation of data value
13. Data_Value_Footnote_Symbol > Symbol that would be used to flag footnotes
14. Data_Value_Footnote > Footnote text
15. Low_Confidence_Limit > Low 95% Confidence Interval value
16. High_Confidence_Limit > High 95% Confidence Interval value
17. Sample_Size > Sample Size
18. Total > Total/Overall breakout category
19. Age(years) > Age(years) breakout category
20. Education > Education breakout category
21. Gender > Gender breakout category
22. Income > Income breakout category
23. Race/Ethnicity > Race/Ethnicity breakout category
24. GeoLocation > Latitude & Longitude to be provided for formatting GeoLocation or Geocode in the format (latitude, longitude)
25. ClassID > Lookup identifier value for Class
26. TopicID > Lookup identifier value for Topic
27. QuestionID > Lookup identifier value for Question
28. DataValueTypeID > Lookup identifier value for Data_Value_type
29. LocationID > Lookup identifier value for Location
30. StratificationCategory1 > Lookup Identification value, such as Age Group, Gender
31. Stratification1 > Data stratified by this value, such as Male, Female, Total
32. StratificationCategoryId1 > Lookup identifier value for Stratification1
33. StratificationID1 > Lookup identifier value for StratificationCategory1
